```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# 1. Load data asli (dataset mentah)
# Pastikan ini file yang sama yang kamu pakai untuk training
df = pd.read_csv("../data/dataset_miniproject.csv") 

# 2. Setup ID dan Target
# Nama kolom harus sama persis dengan di CSV
ids = df['Employee_ID']
y = df['Attrition_Risk_Level'] 

# 3. Lakukan Split Ulang (Hanya untuk mengambil urutan ID)
# PENTING: Pakai random_state=52 (sesuai spreadsheetmu)
X_train, X_test, y_train, y_test, id_train, id_test = train_test_split(
    df, y, ids, 
    test_size=0.4,   # Sesuai config 50:50
    random_state=52, # KUNCI UTAMA
    stratify=y       # PENTING: Cek kode trainingmu dulu. 
                     # Jika saat training kamu pakai stratify=y, disini juga HARUS pakai.
                     # Jika tidak yakin, coba cek kode training lamamu. 
                     # Defaultnya biasanya tidak pakai stratify kecuali ditulis.
)

# 4. Simpan urutan ID yang benar ke file Excel/CSV baru
# Reset index agar rapi
id_test_reset = id_test.reset_index(drop=True)
id_test_reset.to_csv("urutan_id_untuk_analisis_50_50.csv", index=False, header=["Employee_ID_Corrected"])

print("Selesai! File 'urutan_id_untuk_analisis_50_50.csv' sudah dibuat.")
print("Silakan buka file tersebut, copy kolomnya, dan paste ke sheet 'Config-1 (50:50) Analisis'.")
print("\nContoh 5 ID pertama di data Test:")
print(id_test_reset.head())
```
